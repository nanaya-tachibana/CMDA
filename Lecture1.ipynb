```python
import numpy as np
import scipy.fftpack as fft
import matplotlib.pyplot as plt
```

## FFT example

```python
L = 20
n = 128

x = np.linspace(-L/2, L/2, n+1)[:n]

u = np.exp(-x ** 2)
ut = fft.fft(u)
utshift = fft.fftshift(ut)
```

```python
fig, axes = plt.subplots(3, 1, figsize=(6, 12))
axes[0].plot(x, u)
axes[1].stem(abs(ut))
axes[2].stem(abs(utshift))

plt.tight_layout()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

k = 2*np.pi / L * np.hstack([np.arange(0, n / 2), np.arange(-n / 2, 0)])

ax = axes[0, 0]
ax.plot(x, u)
ax.set_xlabel(r'$x$', fontsize=14)
ax.set_ylabel(r'$u=exp(-x^2)$', fontsize=14)

ax = axes[0, 1]
ax.stem(np.real(ut))
ax.set_xlabel(r'$modes$', fontsize=14)
ax.set_ylabel(r'$REAL[U]$', fontsize=14)

ax = axes[1, 0]
ax.stem(np.real(utshift))
ax.set_xlabel(r'$modes$', fontsize=14)
ax.set_ylabel(r'$REAL[fftshift(U)]$', fontsize=14)

ax = axes[1, 1]
ax.plot(fft.fftshift(k), abs(utshift))
ax.set_xlabel(r'$fftshift[k]$', fontsize=14)
ax.set_ylabel(r'$abs[fftshift(U)]$', fontsize=14)

plt.tight_layout()
```

## FFT versus Finite-Difference Differentiation

```python
# Using the FFT to approximate the derivative of sech(x)
f = lambda x: 1 / np.cosh(x)
```

```python
df = lambda x: -f(x) * np.tanh(x)
ddf = lambda x: f(x) - 2 * f(x) ** 3
```

```python
L, n = 20, 128
x = np.linspace(-L/2, L/2, n+1)[:n]

dx = x[1] - x[0]
u = f(x)  # target
ut = fft.fft(u)  # FFT of the target
k = 2*np.pi / L * np.hstack([np.arange(0, n / 2), np.arange(-n / 2, 0)])

# use FFT to calculate the derivatives
ut1 = 1j * k * ut  # first derivative
ut2 = (1j * k) ** 2 * ut  # second derivative
# inverse transform
u1 = np.real(fft.ifft(ut1))
u2 = np.real(fft.ifft(ut2))
# analytic derivative
u1_exact = df(x)
u2_exact = ddf(x)

# finite difference calculation of first derivative
# second order
ux = np.zeros(n)
ux[0] = (-3*u[0] + 4*u[1] - u[2]) / (2*dx)
ux[1:-1] = (u[2:] - u[:-2]) / (2*dx)
ux[-1] = (3*u[-1] - 4*u[-2] + u[-3]) / (2*dx)

# fouth order
ux2 = np.zeros(n)
ux2[0] = (-3*u[0] + 4*u[1] - u[2]) / (2*dx)
ux2[1] = (-3*u[1] + 4*u[2] - u[3]) / (2*dx)
ux2[2:-2] = (-u[4:] + 8*u[3:-1] - 8*u[1:-3] + u[:-4]) / (12*dx)
ux2[-2] = (3*u[-2] - 4*u[-3] + u[-4]) / (2*dx)
ux2[-1] = (3*u[-1] - 4*u[-2] + u[-3]) / (2*dx)
```

```python
plt.plot(x, u1, 'r', x, u1_exact, 'cs', x, u2, 'm', x, u2_exact, 'go')
fig, axes = plt.subplots(1, 3, figsize=(18, 4))
axes[0].plot(x, u1_exact, 'k-', x, u1, 'rx', x, ux, 'bo', x, ux2,'ms', fillstyle='none', markeredgewidth=1.2)
axes[0].set_xlim([-1.15, -0.75])
axes[0].set_ylim([0.47, 0.5])
axes[1].plot(x, u1_exact, 'k-', x, u1, 'rx', x, ux, 'bo', x, ux2,'ms', fillstyle='none', markeredgewidth=1.2)
axes[1].set_xlim([-0.9376, -0.9374])
axes[1].set_ylim([0.49848, 0.49850])
axes[2].plot(x, u1_exact, 'k-', x, u1, 'rx', x, ux, 'bo', x, ux2,'ms', fillstyle='none', markeredgewidth=1.2)
axes[2].set_xlim([-0.9376, -0.9374])
axes[2].set_ylim([0.498487, 0.498488])
```

```python
# Using the FFT to approximate the derivative of tanh(x)
f = lambda x: np.tanh(x)
df = lambda x: 1 / np.cosh(x) ** 2

L, n = 20, 128
x = np.linspace(-L/2, L/2, n+1)[:n]

dx = x[1] - x[0]
u = f(x)  # target
ut = fft.fft(u)  # FFT of the target

# use FFT to calculate the derivatives
ut1 = 1j * k * ut  # first derivative
# inverse transform
u1 = np.real(fft.ifft(ut1))
# analytic derivative
u1_exact = df(x)

fig, ax = plt.subplots(1, 1, figsize=(6, 4))
ax.plot(x, u1, 'r', x, u1_exact, 'bo', fillstyle='none', markeredgewidth=1.2)
ax.set_ylim(-1.5, 1.5)
```

## Application: Radar Detection and Filtering

```python
L, n = 30, 512

t = np.linspace(-L, L, n+1)[:n]
k = (2*np.pi) / (2*L) * np.hstack([np.arange(0, n/2), np.arange(-n/2, 0)])

u = 1 / np.cosh(t)

fig, axes = plt.subplots(3, 1, figsize=(15, 18))
plt.xlabel(r'$t$', fontsize=18)
xlim = [-10, 10]

ax = axes[0]
ax.plot(t, u, 'k')
ax.set_xlim(xlim)
ax.set_title('no noise')
ax.set_ylabel(r'$f(t)$', fontsize=14)

noise = 1  # noise level
ut = fft.fft(u)
noisy_ut = ut + noise * (np.random.randn(n) + 1j*np.random.randn(n))
noisy_u = fft.ifft(noisy_ut)
ax = axes[1]
ax.plot(t, noisy_u, 'k')
ax.set_xlim(xlim)
ax.set_title('noise level = 1')
ax.set_ylabel(r'$f(t)$', fontsize=14)

noise = 10  # greater noise level
ut = fft.fft(u)
noisy_ut = ut + noise * (np.random.randn(n) + 1j*np.random.randn(n))
noisy_u = fft.ifft(noisy_ut)
ax = axes[2]
ax.plot(t, noisy_u, 'k')
ax.set_xlim(xlim)
ax.set_title('noise level = 5')
ax.set_ylabel(r'$f(t)$', fontsize=14)

plt.tight_layout()
```

```python
fig, axes = plt.subplots(2, 1, figsize=(15, 12))

noise = 10  # noise level
ut = fft.fft(u)
noisy_ut = ut + noise * (np.random.randn(n) + 1j*np.random.randn(n))
noisy_u = fft.ifft(noisy_ut)
ax = axes[0]
ax.plot(t, np.abs(noisy_u), 'k')
ax.set_xlim([-30, 30])
ax.set_title('noise level = 10')
ax.set_xlabel(r'$t$', fontsize=14)
ax.set_ylabel(r'$|f(t)|$', fontsize=14)

ax = axes[1]
shifted_noisy_u = fft.fftshift(noisy_u)
ax.plot(fft.fftshift(k), 
        np.abs(shifted_noisy_u) / np.max(np.abs(shifted_noisy_u)), 'k')
ax.set_xlim([-25, 25])
ax.set_title('noise level = 10')
ax.set_xlabel(r'$wavenumber\ k$', fontsize=14)
ax.set_ylabel('$|f(t)|/max(|f(t)|$', fontsize=14)

plt.tight_layout()
```

```python
def gaussian_filter(ut, k, k0=0, bandwidth=0.2):
    """
    ut: numpy array
        FT signal
    k: 1d numpy array
        wavenumber
    k0: int
        center frequency
    bandwidth: float
        filter bandwidth
    """
    return fft.ifft(ut * np.exp(-bandwidth * (k - k0) ** 2))
```

```python
k0, bandwidth = 0, 0.2
shifted_k = fft.fftshift(k)

fig, axes = plt.subplots(3, 1, figsize=(15, 18))

ax = axes[0]
ax.plot(shifted_k, 
        np.abs(shifted_noisy_u) / np.max(np.abs(shifted_noisy_u)), 'k')
ax.plot(shifed_k, np.exp(-bandwidth * (shifed_k - k0) ** 2), 
        linewidth=3, color='black')
ax.set_xlim([-25, 25])
ax.set_xlabel(r'$wavenumber\ k$', fontsize=14)
ax.set_ylabel(r'$|f(t)|/max(|f(t)|$', fontsize=14)

# frequency domain signal
ax = axes[1]
ax.plot(shifted_k, 
        fft.fftshift(np.abs(noisy_ut) * np.exp(-bandwidth * (k - k0) ** 2)), 
        color='black')
ax.set_xlabel(r'$wavenumber\ k$', fontsize=14)
ax.set_ylabel(r'$|f(t)|/max(|f(t)|$', fontsize=14)

# time domain signal
ax = axes[2]
ax.plot(t, np.abs(gaussian_filter(noisy_ut, k, k0=k0, bandwidth=bandwidth)), 
        color='black', linewidth=3)
ax.set_xlabel(r'$t$', fontsize=14)
ax.set_ylabel(r'$|u|$', fontsize=14)
ax.plot(t, u, color='black')
ax.hlines(0.5, t.min(), t.max(), linestyle='--')

plt.tight_layout()
```
